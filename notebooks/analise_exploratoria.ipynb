### Analise Exploratoria

```python
import pandas as pd
import matplotlib as mt
import seaborn as sn

ytGlobal = pd.read_csv("./global_youtube_creator_data_large.csv")

df = pd.DataFrame(ytGlobal)
df['video_id'] = df['video_id'].astype(str)
df['category'] = df['category'].astype("category")
df['language'] = df['language'].astype("category")
df['region'] = df['region'].astype("category")
df['ads_enabled'] = df['ads_enabled'].astype(bool)


```

#### Checagem de qualidade dos Dados

```python
#Tamanho
print("Tamanho do arquivo")
len(df), df.head(3)
print("\n")

#Numero de ID's duplicados
print("Numero de ID's duplicados")
df[df["video_id"].duplicated()].sum()
print("\n")

#numero de nulos por coluna
print("Numero de nulos po coluna")
df.isna().sum().sort_values(ascending=False)
print("\n")
```
